### Visualizing the Data

Cleaned the data to remove contractions, puncuation, and stopwords. Used lemmati

```python
clothing_data['Review Text'] = clothing_data['Review Text'].str.lower()
```

```python
clothing_data['Review Text'] = clothing_data.apply(lambda x : contractions.fix(x['Review Text']), axis = 1)
```

```python
tokenizer = RegexpTokenizer(r'\w+')
review_text = clothing_data['Review Text'].apply(tokenizer.tokenize)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
review_text = review_text.apply(lambda x: [word for word in x if word not in stop_words])
```

```python
with pd.option_context('display.max_colwidth', None):
    print(review_text.head())
```

```python
clothing_data["Text Tokens"] = review_text.apply(lambda x: ' '.join([item for item in x]))
```

```python
positive_reviews = ' '.join([word for word in clothing_data['Text Tokens'].loc[(clothing_data['Rating'] >= 4)]])
neutral_reviews = ' '.join([word for word in clothing_data['Text Tokens'].loc[(clothing_data['Rating'] == 3)]])
negative_reviews = ' '.join([word for word in clothing_data['Text Tokens'].loc[(clothing_data['Rating'] <= 2)]])
```

```python
positive_tokens = nltk.tokenize.word_tokenize(positive_reviews)
neutral_tokens = nltk.tokenize.word_tokenize(neutral_reviews)
negative_tokens = nltk.tokenize.word_tokenize(negative_reviews)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
positive_lem = [lemmatizer.lemmatize(x) for x in positive_tokens]
neutral_lem = [lemmatizer.lemmatize(x) for x in neutral_tokens]
negative_lem = [lemmatizer.lemmatize(x) for x in negative_tokens]
```

```python
positive_freq = FreqDist(positive_tokens)
positive_freq
```

```python
neutral_freq = FreqDist(negative_tokens)
neutral_freq
```

```python
negative_freq = FreqDist(negative_tokens)
negative_freq
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS,
                      collocations=True).generate(positive_reviews)

plt.imshow(wordcloud, interpolation='bilInear')
plt.axis('off')
plt.show()
```
